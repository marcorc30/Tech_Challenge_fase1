# Tipos de Variveis
* categorica
* discreta
* continua

Em estatstica, uma varivel categrica  uma varivel que pode assumir apenas um nmero limitado, e geralmente fixo, de valores possveis, atribuindo cada indivduo ou outra unidade de observao a um determinado grupo ou categoria nominal com base em alguma propriedade qualitativa. 

Em resumo, a diferena entre as variveis quantitativas discretas e contnuas  que as variveis discretas tm valores finitos e limitados, enquanto as variveis contnuas tm valores infinitos e podem ser medidos em escalas infinitamente pequenas.


# Medidas de tendncia central
* Media
* Mediana
    * Coloque em ordem crescente
    * Se a quantidade for impar, a mediana  o valor central
    * Se a quantidade for par, a mediana  soma dos 2 elementos centrais divido por 2
* Moda
    * Frequencia de numero que mais se repete
    * Se no tiver nmero repetido (mesma frequencia de todos os numeros),  um conjunto "amodal"
    
## Em uma distribuio normal, a mdia, mediana e moda tem o mesmo valor
