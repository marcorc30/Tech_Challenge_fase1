# Tipos de Variáveis
* categorica
* discreta
* continua

Em estatística, uma variável categórica é uma variável que pode assumir apenas um número limitado, e geralmente fixo, de valores possíveis, atribuindo cada indivíduo ou outra unidade de observação a um determinado grupo ou categoria nominal com base em alguma propriedade qualitativa. 

Em resumo, a diferença entre as variáveis quantitativas discretas e contínuas é que as variáveis discretas têm valores finitos e limitados, enquanto as variáveis contínuas têm valores infinitos e podem ser medidos em escalas infinitamente pequenas.


# Medidas de tendência central
* Media
* Mediana
    * Coloque em ordem crescente
    * Se a quantidade for impar, a mediana é o valor central
    * Se a quantidade for par, a mediana é soma dos 2 elementos centrais divido por 2
* Moda
    * Frequencia de numero que mais se repete
    * Se não tiver número repetido (mesma frequencia de todos os numeros), é um conjunto "amodal"
    
## Em uma distribuição normal, a média, mediana e moda tem o mesmo valor
